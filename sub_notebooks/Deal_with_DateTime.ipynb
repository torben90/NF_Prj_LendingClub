```python
import numpy as np
import pandas as pd
```

```python
# import the data
df_time = pd.read_pickle('../data/clean_pass_01.pkl')
```

```python
# defining a function for converting month to a number and the feature into datetime
def time_converter(elem):
    month_to_num = {'Jan':'01', 'Feb':'02', 'Mar':'03', 'Apr':'04', 'May':'05', 'Jun':'06',
                    'Jul':'07', 'Aug':'08', 'Sep':'09', 'Oct':'10', 'Nov':'11', 'Dec':'12'}
    temp_elem = elem.split('-')
    val = '-'.join([month_to_num[temp_elem[0]],temp_elem[1]])
    date = pd.to_datetime(val)
    return date
```

```python
# defining the columns which needs to be confirmed
# 
date_column = ['last_credit_pull_d','earliest_cr_line']
for i in date_column:
    df_time[i] = df_time[i].apply(lambda x: time_converter(x))
df_time.info()
```
